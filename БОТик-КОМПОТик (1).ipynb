БОТик-КОМПОТик

Наш ботик умеет выполнять три функции на тематику величайшего напитка наших бабушек - компотика:

 1. Тестик на то, какой ты компот: для истинных любителей девчачих анкет и журналов
 
 2. Помощь с поиском рецепта по ингредиентам (пока у нас в наличии 15 рецептиков компотиков): стоит только ввести игредиенты в КОРРЕКТНОЙ форме и наш ботик выведет на экран наиболее соответствующий им рецептик компотика
 
 3. Помощь с выбором компотика по настроению: нужно выбрать настроение из предложенных и ботик начнет предлагать компотики под него

```python
def print_recipes(name):
    with open('recipes.csv', 'r', encoding = 'utf-8') as infile:
        lst = []
        for line in infile:
            line = line.strip().split(';')
            if name == line[0]:
                n = 1
                for i in line[1:]:
                    if i == '':
                        break
                    a = f'{n}. {i}'
                    lst.append(a)
                    n += 1
    for i in lst:
        print(i) 
option = input("Выбери функцию, введи ее номер:\n 1. Какой ты компотик?\n 2. Рецептик компотика\n 3. Компотик по настроению?\n ")
winter_compots = {'Комфортик-компотик': ["яблоко", "ванильный сахар", "сахар", "вода"],
                 'Компотик «Кислотный рейв»': ["груша", "лимон", "сахар", "вода"],
                  'Компотик «Сливки общества»': ["слива", "сахар", "вода"],
                  'Компотик «Дионис»': ["виноград", "сахар", "вода"],
                  'Айвовый компотик': ["айва", "сахар", "вода"]}
interesting_compots = {'Компотик из Ранеток': ["яблоко", "смородина", "соль", "сахар", "вода"],
                       'Компотик «Питоняшка»': ["груша", "кизил", "лимонная кислота", "сахар", "вода"],
                       'Хэллоуинпотик': ["тыква", "лимон", "сахар", "вода"],
                       'Еловый компотик: для тех, у кого дома из еды только елка': ["ель", "лимон", "сахар", "вода"]}
summer_compots = {'Компотик «Рыжая чикса»': ["абрикос", "апельсин", "вода", "сахар"],
                  '«В рот компот»: очень вкусный': ["яблоко", "персик", "виноград", "апельсин", "лимон", "сахар", "вода"],
                  'Компотик из кринжовника «Мохито»': ["крыжовник", "лайм", "мята", "сахар", "вода"],
                  'Ягодный компотик «Baby-girl»': ["черешня", "клубника", "сахар", "вода"],
                  'Арбузовый компотик «Яркая кокетка»': ["арбуз", "яблоко", "виноград", "ежевика", "варенье", "сахар", "вода"]}
while option.lower()!='нет':
  if option.strip('.') =="1":
    quizz={'После экзамена по макроэкономике ты скорее всего... ':
      ['пойдешь домой плакать','пойдешь пить чай и сплетничать с Матвеевой',
       'пойдешь гулять','пойдешь в кино на фильм ужасов', 'уйдешь в компотный запой (и не только в компотный)'],
            'Какой цвет из перечисленных тебе больше всего нравится?':
              ['розовый', 'оранжевый', 'синий', 'черный', 'зеленый'],
            'Какая профессия из перечисленных тебе больше всего нравится? ':
            ['модель', 'фокусник', 'программист', 'бизнесмен', 'маньяк'],
            'Перед сложной контрольной ты сокрее всего... ':
              ['тревожно забьешь', 'не будешь знать, какой предмет надо учить',
               'подготовишься по дороге в универ', 'будешь судорожно все повторять', 'не будешь переживать, так как и так все знаешь'],
            'Какой стиль в одежде тебе больше всего нравится? ':
            ['элегантный и романтичный', 'спортивный', 'деловой', 'тотал-блэк', 'костюм клоуна'],
            'Какое описание характера тебе больше всего подходит? ':
            ['искренний и обаятельный', 'веселый и жизнерадостный',
             'спокойный и саркастичный', 'серьезный и ответственный', 'апатичный и мрачный'],
              'Какой язык ты учишь? (если ни один из перечисленных, то какой язык тебе больше всего нравится?) ':
            ['французский', 'испанский', 'немецкий', 'китайский', 'питон'],
            'В каком месте ты бы сейчас хотел оказаться? ':
            ['на подиуме', 'на тусовке', 'в кровати', 'на природе', 'в канаве'],
            'Ты хочешь отчислиться из вуза? ':
              ['да, и меня отчислят', 'да, и я отчислюсь', 'да, но я не отчислюсь',
               'нет, но меня отчислят', 'нет, мне все безумно нравится']}
    counter=0
    number=1
    for question in quizz:
        print(question)
        for variant in quizz[question]:
            print(f'{number}. {variant}')
            number+=1
        number=1
        answer=int(input('Выбери один вариант ответа и введи его номер: '))
        counter+=answer
        print('---------')
    if counter <= 18:
        print('Ты малиновый компотик.\n'
        'Ты добрый, искренний и обаятельный человек.\n'
         'Ты очень креативен, часто фантазируешь и создаешь новые идеи.\n'
          'Про тебя могут говорить, что ты слишком часто летаешь в облаках, однако многие просто не улавливают твой мечтательный вайб.\n'
           'Никого не слушай и делай то, что велит твое малиновое сердце.')
    elif counter <= 25:
        print('Ты компотик из абрикосок.\n'
         'Ты очень веселый и общительный человек, легко вливаешься в незнакомую компанию и быстро соглашаешься на любые авантюры.\n'
          'Про тебя говорят, что ты душа компании и генератор хорошего настроения.\n'
           'Не переставай нести свет в эту мрачную жизнь.')
    elif counter <= 32:
        print('Ты сливовый компотик.\n'
         'Ты спокойный и рассудительный.\n'
         'Серьезный и ответственный в работе, но при этом обладаешь отменным чувством юмора, способным сразить наповал любого человека.\n'
           'Счастливый обладатель баланса на чашах work-life весов, поздравляем!')
    elif counter <= 39:
        print('Ты компотик из черной смородины.\n'
         'Аккуратный, педантичный и строгий.\n'
          'На первый взгляд окружающим может показаться, что ты ЧСВ, однако как только люди начинают узнавать тебя ближе,\n'
           'они понимают, что за черной кислой оболочкой прячется красная и сладкая ягодка, настоящая обаяшка.')
    else:
        print('Ты компот из кринжовника (ой крыжовника).\n'
         'Ты слегка нестандартный.\n'
          'Окружающие могут не понимать твоего взгляда на мир и поэтому называть кринжом.\n'
          'Не расстраивайся. Следуй великому правилу: “кто обзывается, тот сам так называется”.')
  elif option.strip('.') == '2':
    name = set(input("Введи ингредиенты через запятую с пробелом в именительном падеже и в единственном числе: ").split(', '))
    rec={}
    for i in winter_compots:
      if len(set(winter_compots[i])&name)>0:
        lis=[]
        lis.append(set(winter_compots[i]))
        lis.append(set(winter_compots[i])&name)
        rec[i] = lis
    for i in summer_compots:
      if len(set(summer_compots[i])&name)>0:
        lis=[]
        lis.append(set(summer_compots[i]))
        lis.append(set(summer_compots[i])&name)
        rec[i] = lis
    for i in interesting_compots:
      if len(set(interesting_compots[i])&name)>0:
        lis=[]
        lis.append(set(interesting_compots[i]))
        lis.append(set(interesting_compots[i])&name)
        rec[i] = lis
    if len(rec) == 0:
      print('Хмм.. Из таких ингредиентов ты не приготовишь компот. Иди в магазин')
    else:
        rec1 = sorted(rec.items(), key = lambda x:(-len(x[1][1]), len(x[1][0])))
        print(rec1[0][0])
        print(f'Ингредиенты: {", ".join(rec1[0][1][0])}')
        recipe = print_recipes(rec1[0][0])
  elif option.strip('.') == '3':
    mood = input("Какой ты сегодня: 1.тутти-фрутти, 2.отмороженный, 3.загадочный? Введи номер варианта: ")
    if mood.lower() == "1":
      for k in summer_compots:
        print(f'Сегодня ты {k}')
        question = input('Всё ок?: Да или нет')
        if question.lower() == 'да':
            question2 = input('Хочешь рецептик?')
            if question2.lower() == 'да':
                recipe = print_recipes(k)
            else:
                print('Приятного компотопития, дорогой/ая питоняшка!')
            break
      if question.lower() == 'нет':
         print('Капец ты привереда, нам больше нечего тебе предложить(((')
    elif mood.lower() == "2":
      for h in winter_compots:
        print(f'Сегодня ты {h}')
        question = input('Смэш? Да или нет')
        if question.lower()=='да':
            question2 = input('Хочешь рецептик?')
            if question2.lower() == 'да':
                recipe = print_recipes(h)
            else:
                print('Приятного компотопития, дорогой/ая питоняшка!')
            break
      if question.lower()=='нет':
          print('Капец ты привереда, нам больше нечего тебе предложить(((')
    elif mood.lower()=='3':
      for i in interesting_compots:
        print(f'Сегодня ты {i}')
        question=input('Ям? Да или нет? ')
        if question.lower()=='да':
            question2 = input('Хочешь рецептик?')
            if question2.lower() == 'да':
                recipe = print_recipes(i)
            else:
                print('Приятного компотопития, дорогой/ая питоняшка!')
            break
      if question.lower()=='нет':
          print('Капец ты привереда, нам больше нечего тебе предложить(((')
    else:
      print('Ю шур бейби герл? Наш ботик пока не умеет принимать такие вайбы(((')
  else:
    print('Я такого не умею(((')
  question=input('Продолжить? да или нет? ')
  if question.lower() == 'нет':
    print('До следующего компотопития, питоняшка!)')
    break
  elif question.lower() == 'да':
    option = input("Выбери функцию, введи ее номер:\n 1. Какой ты компотик?\n 2. Рецептик компотика\n 3. Компотик по настроению?\n ")
  else:
      print('Мы тебя не понимаем( Попробуй сначала')
      option = input("Выбери функцию, введи ее номер:\n 1. Какой ты компотик?\n 2. Рецептик компотика\n 3. Компотик по настроению?\n ")
```
