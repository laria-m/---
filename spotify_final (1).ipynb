# **Выбросы**

Для удаления выбросов воспользуемся **методом 1.5IQR**


*Мы решили не удалять выбросы по количественным признакам, описывающим аспекты звучания и стиль песен, так как они имеют исследовательскую ценность.*





```python
df_new=df # чтобы не потерять данные из оригинального датасета, создаем новый

def del_vib(x): # функция, которая облегчаяет удаление выбросов по признаку x
    iqr = df[x].quantile(0.75) - df[x].quantile(0.25)
    bt = df[x].quantile(0.25) - 1.5*iqr
    top = df[x].quantile(0.75) + 1.5*iqr
    return df_new[(df_new[x] >= bt) & (df[x]<= top)]

for i in ['artist_count', 'in_spotify_playlists', 'streams', 'in_apple_playlists',
          'in_deezer_playlists']:
    df_new=del_vib(i)

df_new # в обновленном датафрейме представлены данные без выбросов
```

# **Тест на сопряженность (независимость) 𝜒2**

Мы предполагаем, что настроение песни (лад) может зависеть от времени года, проверим так ли это с помощью хи-квадрат тестирования:

Ho: Лад песни не зависит от времени года

H1: Лад песни зависит от времени года

```python
def season(x): #функция, которая распределяет песни по временам года по месяцу релиза
    if x in [1,2,12]:
        return 'Winer'
    elif x in [3,4,5]:
        return 'Spring'
    elif x in [6,7,8]:
        return 'Summer'
    else:
        return 'Autumn'

df_new['Season'] = df_new['released_month'].apply(season)
```

```python
meow=df_new.groupby('Season')['mode'].value_counts().unstack()
meow #наша таблица сопряженности
```

```python
from scipy import stats
from scipy.stats import chi2_contingency

result = chi2_contingency(meow.values) #проводим само хи-тестирование
result
```

**pvalue=8.341%** > 5% уровня значимости, а значит, нулевую гипотезу нельзя отвергнуть: вероятно, что между временем года и настроением песни нет зависимости.

Однако, при большем уровне значимости в 10%, мы отвергаем нулевую гипотезу и тогда по гипотезе H1: связь между временем года и настроением песни есть.

**Вывод: между временем года и ладом песни есть зависимость только при принятии достаточно великой вероятности допустить ошибку первого рода.**

# **Одновыборочный t-тест**

В среднем уровень танцевальности всех песен составляет около 65%, мы предполагаем, что песни топа Spotify имеют примерно такой же уровень танцевальности или выше, так как популярные песни это часто тренды соцсетей и их включают на вечеринках.

Ho: средний уровень танцевальности популярных песен равен 65%

H1: средний уровень танцевальности популярных песен больше 65%

```python
mu0 = 65
result = stats.ttest_1samp(a = df_new['danceability_%'], popmean = mu0, alternative = 'greater')
result
```

**pvalue=0.000** < любого адекватного уровня значимости, значит, мы отвергаем нулевую гипотезу.

# **Эконометрический бонус**

Проведем тест Чоу

**H0: структурного разрыва данных нет**, то есть коэффициенты регрессии для двух подвыборок одинаковы

**H1: структурный разрыв есть**

```python
df1=df_new[df_new['released_year'] < 2019] #рассмотрим треки, выпущенные в период до ковида и после
X1=df1[predicts] #predicts - список признаков, определенный ранее
X1=sm.add_constant(X1) 
y1=df1['sqrt_streams']

df2=df_new[df_new['released_year'] >= 2019]
X2=df2[predicts]
X2=sm.add_constant(X2) 
y2=df2['sqrt_streams']

model1 = sm.OLS(y1, X1).fit() #строим линейные регрессии по подвыборкам
model2 = sm.OLS(y2, X2).fit()

rss = model.ssr #ищем суммы квадратов остатков
rss1 = model1.ssr
rss2 = model2.ssr
k=X.shape[1]
n=len(y)

F =((rss - rss1 - rss2)/k)/((rss1 + rss2)/(n - 2*k)) #проводим тестирование
Fcr=stats.f.ppf(1-0.05, k, n-2*k) #примем альфу - 5%

print(f'F = {F}, Fкритическая = {Fcr}')
```

Так как F > Fкритической, мы отвергаем нулевую гипотезу: структурный разрыв данных есть.

**Вывод: у нас неоднородная выборка в контексте линейной регрессии, если рассматривать песни, периоды выпуска треков до ковида и после.**

**Методологические проблемы:**

1) непропорциональность подвыборок, так как большая часть трендовых песен, выпущена в 2022 и 2023 - можно было выбрать рандомно треки из второго периода

2) сложно определить размеры подвыборок и по какому принципу их делить - можно предположить исходя из информации, которая доступна из новостей или исследований, примерное время структурного сдвига
